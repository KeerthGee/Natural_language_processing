# Tokenization:
    
    Breaking down into smaller units to work with. 
    
Types:
    
    1.Sentence Tokenization - spliting as a sentences.
    
    2. Word Tokenization - Spliting as a words.
    

# Sentence Tokenization - Spliting as a sentences

```python
from nltk.tokenize import sent_tokenize

text = """Hello Mr.Bing, hope you are doing good.
By the way I was planning on visiting you the next week"""

print('Original text:')
print('='*79)
print(text)
print('='*79)
print()
tokenizes_sent = sent_tokenize(text)
print('After sentence Tokenization: \n',tokenizes_sent)
print()
print('='*79)
print('No. of sentences:\t',len(tokenizes_sent))
```

# Word Tokenization - Spliting as a word

```python
from nltk.tokenize import word_tokenize
 
print('Original text:')
print('='*79)
print(text)
print('='*79)
print()
 
token_word=word_tokenize(text)
 
print('After word Tokenization:\n',token_word)
print('='*79)
print('No.of words:\t',len(token_word))
print('='*79)
 
for i in token_word:
    print('word  : ',i,'&','Length   :',len(i))
```

# Frequency Distribution

```python
from nltk.probability import FreqDist
freq_dist = FreqDist(token_word)
print(type(freq_dist))
print(freq_dist)
```

```python
freq_dist.most_common()
```

```python
import matplotlib.pyplot as plt
freq_dist.plot(27,cumulative = False)
plt.show()
```

# Stopwords

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
print('Total No. of Stopwords:\t',len(stop_words))
print('='*79)
print('The Stop words are:\n')
print(stop_words)
```

# Removing stopwords

```python
filtered_tokens=[]
for w in token_word:
    if w not in stop_words:
        filtered_tokens.append(w)
print('Length of words:\t',len(token_word))
print('-'*79)
print('Tokenized words with stopwords:\n\n\t',token_word)
print('='*79)
print('Length after remove stopwords:\t',len(filtered_tokens))
print('-'*79)
print('\n After removing stopwords- words are:\n\t',filtered_tokens)
print('='*79)
```

# Stemming:
    
    1. Text Normalization technique
    
    2. Used to trim suffix the words to their base form
    
    3. Map different forms of word back to its common root to simplify analysis or improve text matching

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

stemmed_words=[]

for w in filtered_tokens:
    stemmed_words.append(ps.stem(w))
    
print('Length of words:\t',len(filtered_tokens))
print('-'*79)
print('Filtered words:\n\n\t',filtered_tokens)
print('='*79)
print('Length after Stemmer:\t',len(stemmed_words))
print('-'*79)
print('\n After stemming- words are:\n\t',stemmed_words)
```

# Lemmatization:
    
    Lemma means base form a word
    
    Aims to reduce the words to their basse form
    
    Unlike Stemming, Lemmatization considers the context and part of speech of a word to determine its lemma
    
    It ensures that the resulting lemma is a valid word

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer

#Initialize the wordnet Lemmatizer
lemma=WordNetLemmatizer()

#Prompt to take take input string
text = input('Enter a sentence:\n\n\t')

#split as word tokens
words = word_tokenize(text)
print('='*79)
print('Tokens of word:\n',words)
print('Length:\t',len(words))
print('*'*79)

#Apply Lemmatization to each word

lemma_words = [lemma.lemmatize(word,pos = 'v') for word in words]
print('-'*79)
print('Original text:\n',text)
print('='*79)
print('Lemmatized words:\n',lemma_words)
```

```python
import nltk
from nltk import pos_tag

#prompt to take input string
text = input('Enter a sentences:\n\n\t')

#split as word tokens
words = word_tokenize(text)
print('='*79)
print('Tokens of word:\n',words)
print('Length: \t',len(words))
print('*'*79)

#Apply POS tag to each word

pos_tags=pos_tag(words)
print('-'*79)
print('Original text:\n',text)
print('='*79)

for w in pos_tags:
    print('='*40)
    print('POS Tagging',w)
    print('='*40)
#print('Pos Tags:\n',pos_tags)
```

# Filter out Punctuation

```python
import nltk
from nltk import pos_tag

#prompt to take input string
text = input('Enter a sentences:\n\n\t')

#split as word tokens
tokens = word_tokenize(text)
print('='*79)
print('Length:\t',len(tokens))
print('Tokens of word:\n',tokens)
print('*'*79)

#Remove all tokens that are not alphabetic

words = [word for word in tokens if word.isalpha()]

print('-'*79)
print('Orgiginal text:\n',text)
print('='*79)

print('After Removing Punctuation:\n')
print(words,len(words))
```

# Word counts with CountVectorizer

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

documents = input('Text paragraph please:\n\n\t')

sent_tokens = sent_tokenize(documents)
print('Sentences as token:')
for sent in sent_tokens:
    print(sent)
    
#Initialize countvectorizer
vector = CountVectorizer()

#Fit and transform the documents
x = vector.fit_transform(sent_tokens)

#get the feature names(words)
feature_names = vector.get_feature_names_out()

#Display the matrix of token counts
print('Feature names:\n',feature_names)
print('='*79)
print('Token Counts Matrix:')
print(x.toarray())
```
